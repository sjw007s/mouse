```python
def k_means(total_data, enviroment):
    for k in range(number_mouse): #train center
        if enviroment.flag ==1: 
            enviroment.flag=0
            #seed_x = np.random.randint(len(total_data),size=number_mouse)
            seed_initial = total_data[0:500*number_mouse:500].copy()


        arrange_ = np.zeros((len(total_data),number_mouse))
        for j in range(number_mouse):
            arrange_[:,j] = np.power((seed_initial[j,0]-total_data[:,0]),2) + np.power((seed_initial[j,1]-total_data[:,1]),2)
        arrange_group = np.argmin(arrange_,axis=1)
        #print(arrange_group.shape)
        for j in range(number_mouse):
            try:
                temp_arrange_index = np.where(j==arrange_group)
                temp_len = len(temp_arrange_index[0])
                temp_average_y = np.sum(total_data[temp_arrange_index][:,0])/temp_len
                temp_average_x = np.sum(total_data[temp_arrange_index][:,1])/temp_len
                #print(seed_initial)
                #if abs(seed_initial[j][0]-int(temp_average_y)) <500 and \
                #       abs(seed_initial[j][1]-int(temp_average_x))<500: #if Error is big than predted.
                seed_initial[j][1] = int(temp_average_x)
                seed_initial[j][0] = int(temp_average_y)
                #else:
                #    print("악")
                #if frame_range[seed_initial[j,0],seed_initial[j,1]] != 255:
                #    seed_x = np.random.randint(len(total_data),size=1)
                #    seed_initial[j] = total_data[seed_x].copy()
            except:
                pass
```
